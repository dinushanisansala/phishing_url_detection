### Common Functions


```python
def load_data(file_path):
    """
    Loads a CSV file into a pandas DataFrame with error handling.

    Parameters:
    file_path (str): The path to the CSV file.

    Returns:
    pd.DataFrame: The loaded DataFrame if successful, None otherwise.
    """
    try:
        # Attempt to load the dataset
        data = pd.read_csv(file_path)
        return data
    except FileNotFoundError:
        # Handle case where the file does not exist
        print(f"Error: The file at '{file_path}' was not found.")
    except pd.errors.EmptyDataError:
        # Handle case where the file is empty
        print(f"Error: The file at '{file_path}' is empty.")
    except pd.errors.ParserError:
        # Handle case where there is a parsing error
        print(f"Error: There was a parsing error while reading the file at '{file_path}'.")
    except Exception as e:
        # Handle any other unexpected errors
        print(f"An unexpected error occurred: {e}")


def create_directory(dir_name):
    """
    Create a directory inside the parent directory of the current working directory.

    Parameters:
    dir_name (str): The name of the directory to create.
    
    Returns:
    str: The path of the created directory or an error message if the creation fails.
    """
    try:
        # Validate input
        if not isinstance(dir_name, str) or not dir_name:
            raise ValueError("The directory name must be a non-empty string.")

        # Define the current directory path
        current_directory = os.getcwd()

        # Define the parent directory path
        parent_directory = os.path.abspath(os.path.join(current_directory, os.pardir))

        # Define the output directory within the parent directory
        output_dir = os.path.join(parent_directory, dir_name)

        # Create the directory if it doesn't exist
        os.makedirs(output_dir, exist_ok=True)

        print(f"Directory created at: {output_dir}")
        return output_dir

    except OSError as e:
        print(f"Error creating directory '{dir_name}': {e}")
        return None

    except ValueError as e:
        print(f"Invalid input: {e}")
        return None

def get_file_path(dir_name, file_name=None):
    """
    Constructs the full path to a file in the specified directory under the parent directory.

    Parameters:
    - dir_name (str): The name of the directory.
    - file_name (str): The name of the file (optional).

    Returns:
    - str: The full path to the file in the specified directory.
    """
    # Validate input
    if not isinstance(dir_name, str) or not dir_name:
        raise ValueError("The directory name must be a non-empty string.")

    current_directory = os.getcwd()
    # Define the parent directory path
    parent_directory = os.path.abspath(os.path.join(current_directory, os.pardir))
    # Define the target directory path
    target_directory = os.path.join(parent_directory, dir_name)

    try:
        # Create the directory if it doesn't exist
        os.makedirs(target_directory, exist_ok=True)
    except OSError as e:
        print(f"Error creating directory '{dir_name}': {e}")
        return None

    # Define the file path relative to the target directory
    file_path = os.path.join(target_directory, file_name) if file_name else target_directory
    return file_path

def load_model_and_predictors(model_name, directory):
    """
    Load a model and its predictors from the specified directory.

    Parameters:
    - model_name (str): The name of the model to load.
    - directory (str): The directory where the model and predictors are saved.

    Returns:
    - model: The loaded model object.
    - predictors: The list of predictors/features.
    """
    model_file_path = os.path.join(directory, f'{model_name}_model.joblib')
    predictors_file_path = os.path.join(directory, f'{model_name}_predictors.joblib')
    
    # Load model
    model = joblib.load(model_file_path)
    
    # Load predictors
    predictors = joblib.load(predictors_file_path)
    
    return model, predictors
```

Based on the characteristics of the dataset:

In our dataset, the target variable of interest is `medicalCost`, which represents the healthcare expenditures of individuals.

**Supervised, Unsupervised, or Semi-supervised Learning?**

Supervised learning entails training a model on labeled data, where both input features (such as age, sex, BMI, children, smoker status, and region) and corresponding target labels (medical costs) are provided. This enables the model to learn the mapping function from the input features to the output, thereby facilitating predictions for new data points.

**Classification, Clustering, Regression Analysis, or Other?**

Given that `medicalCost` is a continuous numerical variable, the appropriate machine learning task in this scenario is **regression analysis**. Regression models are designed to predict continuous outputs, making them well-suited for estimating medical costs based on the provided demographic and lifestyle information.

**Why Regression Analysis?**
- Regression is specifically designed to handle prediction tasks where the target is continuous.
- Regression models provide coefficients that can be interpreted to understand the impact of each feature (age, sex, BMI, etc.) on medical costs. This interpretability is crucial in domains like healthcare where understanding relationships is important.
- Regression models use metrics such as Mean Squared Error (MSE), Mean Absolute Error (MAE), or R-squared (coefficient of determination) to evaluate model performance, which are suitable for assessing predictions against continuous targets.

```python
# Descriptive statistics Analysis
print("Information about the dataset:")
display(df.describe())

# Analyze descriptive statistics by grouping each categorical column for numerical columns only
for cat in categorical_features:
    print(f"\nDescriptive Analysis by '{cat}' for medicalCost")
    display(df.groupby(cat)['medicalCost'].describe().reset_index())
    print(f"{'#' * 50}")
```

```python
# Columns to detect outliers
outlier_columns = ['bmi', 'medicalCost']

# Dictionary to store outliers
outliers = {}

# Detect outliers using both Z-score and IQR methods
for col in outlier_columns:
    # Z-score method
    mean = np.mean(df[col])
    std = np.std(df[col])
    z_scores_threshold = 3
    z_scores = np.abs((df[col] - mean) / std)
    z_scores_outliers = df[z_scores > z_scores_threshold]
    outliers[f"{col}_z_score"] = z_scores_outliers
    
    # IQR method
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    iqr_lower_threshold = Q1 - 1.5 * IQR
    iqr_upper_threshold = Q3 + 1.5 * IQR
    iqr_outliers = df[(df[col] < iqr_lower_threshold) | (df[col] > iqr_upper_threshold)]
    outliers[f"{col}_iqr"] = iqr_outliers

# Print all detected outliers
for key, value in outliers.items():
    print(f"Potential outliers in '{key}':")
    print(value)
    print(f"{'#' * 50}")
```

- Correlation Analysis: Compute the correlation matrix of numeric features, typically required for correlation analysis.

- Heatmap Visualization: Visualize the correlation matrix as a heatmap using Seaborn's heatmap() function.

- Correlation with Target Variable: Extract correlation values between all features and the target variable 'medicalCost', then sort and print these values.

```python
# Encode categorical variables using one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_features)

# Selecting encoded columns and the target variable
encoded_columns = [col for col in df_encoded.columns if col.startswith('sex_') or col.startswith('smoker_') or col.startswith('region_')]
numerical_features = numerical_features + encoded_columns 
```

```python
# Compute the correlation matrix
correlation_matrix = df_encoded[numerical_features].corr()

# Print correlation matrix
print("Correlation Matrix:")
#print(correlation_matrix)

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Extract the correlation values with 'medicalCost'
medicalCost_corr = correlation_matrix['medicalCost'].sort_values(ascending=False)

# Print the correlation values with 'medicalCost'
print("Correlation with 'medicalCost':")
print(medicalCost_corr)

# Visualize the correlation with 'medicalCost' as a bar plot
plt.figure(figsize=(10, 8))
sns.barplot(x=medicalCost_corr.index, y=medicalCost_corr.values, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Correlation with MedicalCost')
plt.ylabel('Correlation coefficient')
plt.xlabel('Features')
plt.show()
```

### Simple Linear Regression Models for highest Corelation Predictors

This step fits linear regression models for defined predictors individualy for prediction.

- Initializes separate linear regression models for defined predictors.
- Fits each model using training data and makes predictions on test data.
- Stores model details including prediction time.
- Visualizes scatter plots with regression and ideal lines for each predictor against 'medicalCost'.
- Generates bar plots for Mean Squared Error (MSE), R-squared (R2), and prediction time for each predictor's model.

```python
# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_encoded[predictors], df_encoded[target], test_size=0.2, random_state=42)

for predictor in predictors:
    model = LinearRegression()
    
    # Fit the model and measure time taken
    start_time = time.time()
    model.fit(X_train[[predictor]], y_train)
    time_taken = time.time() - start_time
    
    # Make predictions
    y_pred = model.predict(X_test[[predictor]])
    
    # Store model and predictions
    models[predictor] = {'model': model,'predictors':predictor, 'predictions': y_pred,'trains': y_test, 'time_taken': time_taken}


# Plotting scatter plots and regression lines for each predictor in a row
plt.figure(figsize=(16, 5))
for idx, predictor in enumerate(predictors):
    plt.subplot(1, 3, idx + 1)
    plt.scatter(X_test[predictor], y_test, color='blue', label='Actual')
    plt.plot(X_test[predictor], models[predictor]['predictions'], color='red', linewidth=2, label='Predicted')
    plt.title(f"Linear Regression (regression lines): {predictor} vs {target}")
    plt.xlabel(predictor)
    plt.ylabel(target)
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()

# Plotting scatter plots and ideal lines for each predictor in a row
plt.figure(figsize=(16, 5))
for idx, predictor in enumerate(predictors):
    plt.subplot(1, 3, idx + 1)
    plt.scatter(X_test[predictor], y_pred, color='blue', label='Actual')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='green', linestyle='--', linewidth=2, label='Ideal line')
    plt.title(f"Linear Regression (ideal Line): {predictor} vs {target}")
    plt.xlabel(predictor)
    plt.ylabel(target)
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()

# Plotting performance metrics as bar plots
plt.figure(figsize=(12, 6))

# Mean Squared Error (MSE) plot
plt.subplot(1, 3, 1)
plt.bar(models.keys(), [mean_squared_error(y_test, models[predictor]['predictions']) for predictor in predictors], color=['darkblue', 'darkred', 'darkgreen'])
plt.title('Mean Squared Error (MSE)')
plt.ylabel('MSE')
plt.grid(True)

# R-squared (R2) plot
plt.subplot(1, 3, 2)
plt.bar(models.keys(), [r2_score(y_test, models[predictor]['predictions']) for predictor in predictors], color=['darkblue', 'darkred', 'darkgreen'])
plt.title('R-squared (R2)')
plt.ylabel('R2')
plt.grid(True)

# Time taken plot
plt.subplot(1, 3, 3)
plt.bar(models.keys(), [models[predictor]['time_taken'] for predictor in predictors], color=['darkblue', 'darkred', 'darkgreen'])
plt.title('Time Taken (Seconds)')
plt.ylabel('Time')
plt.grid(True)

plt.tight_layout()
plt.show()

# Evaluate models and print performance metrics
for predictor, model_data in models.items():
    model = model_data['model']
    y_pred = model_data['predictions']
    time_taken = model_data['time_taken']
    
    # Print model performance metrics
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f"Linear Regression Model for predictor: {predictor}")
    print("=" * 50)
    print(f"Coefficient: {model.coef_[0]:.2f}")
    print(f"Intercept: {model.intercept_:.2f}")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"R-squared (R2): {r2:.2f}")
    print(f"Time Taken: {time_taken:.5f} seconds")
    print("\n")
```

```python
# Selecting predictors and target variable
predictors = ['smoker_yes', 'age', 'bmi']
target = 'medicalCost'

# Creating X_selected with selected predictors
X_selected = df_encoded[predictors]

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, df_encoded[target], test_size=0.2, random_state=42)

# Initialize linear regression model
model = LinearRegression()

# Fit the model and measure time taken
start_time = time.time()
model.fit(X_train, y_train)
time_taken = time.time() - start_time

# Make predictions
y_pred = model.predict(X_test)

models['three_predictors'] = {'model': model,'predictors':predictors, 'predictions': y_pred,'trains': y_test, 'time_taken': time_taken}


# Plotting scatter plot of actual vs predicted values with predictor line
plt.figure(figsize=(12, 6))

# Scatter plot of actual vs predicted values
plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2, label='Ideal line')
plt.title('Actual vs Predicted Medical Costs - Multivariate Regression Model with highest three correlation predictors')
plt.xlabel('Actual Medical Costs')
plt.ylabel('Predicted Medical Costs')
plt.grid(True)
plt.legend()

# Plotting predictor line
plt.subplot(1, 2, 2)
plt.bar(predictors, model.coef_)
plt.axhline(0, color='grey', linewidth=0.8)
plt.title('Coefficients of Predictors')
plt.xlabel('Predictors')
plt.ylabel('Coefficient Value')
plt.grid(True)

plt.tight_layout()
plt.show()
# Evaluate model and print performance metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Multivariate Regression Model with highest three corrilation predictors")
print("=" * 50)
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2): {r2:.4f}")
print(f"Time Taken: {time_taken:.5f} seconds")
print("\n")
```

#### Function to Evaluate KMeans clustering model using default hyperparameters

This function evaluates a KMeans clustering model using defined hyperparameters. It preprocesses the data, splits it into training and testing sets, fits the KMeans model to the training data, makes predictions on the test set, and calculates performance metrics specific to clustering (inertia, silhouette score, Davies-Bouldin index). It does not use traditional cross-validation due to the nature of clustering. The evaluation results are stored in a dictionary (model_result) which is returned by the function.

```python
def evaluate_kmeans_models(model_name, classifier, df, target_column, test_size=0.2):
    """
    Evaluate KMeans clustering model using default hyperparameters.

    Parameters:
    - model_name (str): The name of the machine learning model.
    - classifier: KMeans clustering model.
    - df (pd.DataFrame): Input dataframe containing features and target.
    - target_column (str): Name of the target column in the dataframe.
    - test_size (float): Proportion of the dataset to include in the test split (default=0.2).

    Returns:
    - model_result (dict): Dictionary containing evaluation results for the model.
    """
    model_result = {}  # Dictionary to store evaluation results

    # Identify numerical and categorical columns
    numerical_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_features = df.select_dtypes(include=['object']).columns.tolist()
    
    # Remove target column from the features lists
    if target_column in numerical_features:
        numerical_features.remove(target_column)
    if target_column in categorical_features:
        categorical_features.remove(target_column)
        
    # Split the dataset into training and testing sets
    X = df.drop(target_column, axis=1)
    y = df[target_column]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Define the preprocessing steps for numerical and categorical features
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numerical_features),
            ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
        ]
    )

    # Fit and transform the preprocessor on X_train
    X_train_preprocessed = preprocessor.fit_transform(X_train)
    X_test_preprocessed = preprocessor.transform(X_test)

    try:
        # For KMeans, preprocess the data outside the pipeline and fit the classifier directly
        y_train = y_train.apply(str)
        y_test = y_test.apply(str)
                
        # For KMeans, preprocess the data outside the pipeline and fit the classifier directly
        start_time = time.time()
        classifier.fit(X_train_preprocessed)
        time_taken = time.time() - start_time
        
        # Make predictions
        y_pred = classifier.predict(X_test_preprocessed)

        # Convert sparse matrices to dense arrays for KMeans-specific metrics
        if hasattr(X_test_preprocessed, "toarray"):
            X_test_dense = X_test_preprocessed.toarray()
        else:
            X_test_dense = X_test_preprocessed

        # Calculate performance metrics specific to KMeans
        inertia = classifier.inertia_
        silhouette = silhouette_score(X_test_dense, y_pred)
        davies_bouldin = davies_bouldin_score(X_test_dense, y_pred)

        # KMeans does not use cross-validation in the same sense, so it's set to None
        cv_scores = None

        # Store the results in the dictionary
        model_result = {
            'estimator': classifier,
            'inertia': inertia,
            'silhouette_score': silhouette,
            'davies_bouldin_index': davies_bouldin,
            'time_taken': time_taken,
            'cv_scores':cv_scores
        }

    except Exception as e:
        print(f"Error evaluating model 'KMeans': {e}")
        model_results = {'error': str(e)}

    return model_result

```

#### Function to Evaluate KMeans clustering model using default hyperparameters

This function evaluates a deep neural network (DNN) model using default hyperparameters. It preprocesses the data, splits it into training and testing sets, initializes the DNN model using the provided classifier, constructs a pipeline including preprocessing steps, fits the pipeline to the training data, evaluates the model on the test set, calculates performance metrics (accuracy, confusion matrix, classification report), and computes cross-validation scores using Stratified K-Fold. The evaluation results are stored in a dictionary (model_result) which is returned by the function.

```python
def evaluate_dnn_model(model_name, classifier_fn, df, target_column, cross_validation=5, test_size=0.2):
    """
    Evaluate a machine learning model using default hyperparameters.

    Parameters:
    - model_name (str): The name of the machine learning model.
    - classifier: Defined Deap Nural Network model.
    - df (pd.DataFrame): Input dataframe containing features and target.
    - target_column (str): Name of the target column in the dataframe.
    - cross_validation (int): Number of cross-validation folds (default=5).
    - test_size (float): Proportion of the dataset to include in the test split (default=0.2).

    Returns:
    - model_result (dict): Dictionary containing evaluation results for the model.
    """
    model_result = {}  # Dictionary to store evaluation results

    # Identify numerical and categorical columns
    numerical_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_features = df.select_dtypes(include=['object']).columns.tolist()
    
    # Remove target column from the features lists
    if target_column in numerical_features:
        numerical_features.remove(target_column)
    if target_column in categorical_features:
        categorical_features.remove(target_column)
        
    # Split the dataset into training and testing sets
    X = df.drop(target_column, axis=1)
    y = df[target_column]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Define the preprocessing steps for numerical and categorical features
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numerical_features),
            ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
        ]
    )

    try:
        # Fit and transform the preprocessor on X_train
        X_train_preprocessed = preprocessor.fit_transform(X_train)
        X_test_preprocessed = preprocessor.transform(X_test)

        # Initialize the classifier
        input_dim = X_train_preprocessed.shape[1]
        classifier = KerasDNN(input_dim=input_dim, epochs=50, batch_size=32, learning_rate=0.001)

        # Define the pipeline with preprocessing and the classifier
        pipeline = Pipeline([
            ('preprocessor', preprocessor),
            ('classifier', classifier)
        ])
        
        # Fit the pipeline
        start_time = time.time()
        pipeline.fit(X_train, y_train)
        time_taken = time.time() - start_time

        # Make predictions
        y_pred = pipeline.predict(X_test)

        # Calculate performance metrics
        accuracy = accuracy_score(y_test, y_pred)
        conf_matrix = confusion_matrix(y_test, y_pred)
        class_report = classification_report(y_test, y_pred)

        # Calculate cross-validation score and handle exceptions
        cv_scores = []
        skf = StratifiedKFold(n_splits=cross_validation, random_state=42, shuffle=True)
        for train_index, val_index in skf.split(X_train_preprocessed, y_train):
            X_train_cv, X_val = X_train_preprocessed[train_index], X_train_preprocessed[val_index]
            y_train_cv, y_val = y_train.iloc[train_index], y_train.iloc[val_index]
            classifier.fit(X_train_cv, y_train_cv)
            cv_scores.append(classifier.model.evaluate(X_val, y_val)[1])  # Append validation accuracy score

        cv_scores = np.array(cv_scores)

        # Store the results in the dictionary
        model_result = {
            'estimator': pipeline,
            'accuracy': accuracy,
            'confusion_matrix': conf_matrix,
            'classification_report': class_report,
            'cross_val_score': cv_scores,
            'time_taken': time_taken
        }
    except Exception as e:
        print(f"Error evaluating model '{model_name}': {e}")
        model_result = {'error': str(e)}

    return model_result
```

#### Function to Evaluate multiple models with optional hyperparameters tunning feature

This function evaluates multiple machine learning models specified in ml_models using default hyperparameters. It preprocesses the data, splits it into training and testing sets, iterates over each model, constructs pipelines including preprocessing steps and the classifier, fits each pipeline to the training data, evaluates each model on the test set, calculates performance metrics (accuracy, confusion matrix, classification report), and computes cross-validation scores using Stratified K-Fold. Optionally, it can perform randomized hyperparameter search using param_distributions. The evaluation results for each model are stored in a dictionary (model_results) which is returned by the function.

```python
def evaluate_ml_models(ml_models, df, target_column, cross_validation=5, test_size=0.2, param_distributions=None, n_iter=10):
    """
    Evaluate multiple machine learning models using default hyperparameters and optionally perform randomized hyperparameter search.

    Parameters:
    - ml_models (dict): Dictionary containing classifiers for each model.
    - df (pd.DataFrame): Input dataframe containing features and target.
    - target_column (str): Name of the target column in the dataframe.
    - cross_validation (int): Number of cross-validation folds (default=5).
    - test_size (float): Proportion of the dataset to include in the test split (default=0.2).
    - param_distributions (dict or None): Dictionary with parameters names (string) as keys and distributions
                                          or lists of parameters to try. If None, default hyperparameters are used.
    - n_iter (int): Number of parameter settings that are sampled. Defaults to 10.

    Returns:
    - model_results (dict): Dictionary containing evaluation results for each model.
    """
    model_results = {}  # Dictionary to store evaluation results

    # Identify numerical and categorical columns
    numerical_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_features = df.select_dtypes(include=['object']).columns.tolist()
    
    # Remove target column from the features lists
    if target_column in numerical_features:
        numerical_features.remove(target_column)
    if target_column in categorical_features:
        categorical_features.remove(target_column)

    # Iterate over each model
    for model_name, model_info in ml_models.items():
        print("#" * 50)
        print(f"{model_name} Model Evaluation")
        print("#" * 50)
        
        try:
            # Extract the classifier from the dictionary
            classifier = model_info['classifier']
                
            if model_name == 'DNN':
                model_results[model_name] = evaluate_dnn_model(model_name, classifier, df, target_column)
            elif model_name == 'KMeans':
                model_results[model_name] = evaluate_kmeans_models(model_name, classifier, df, target_column)
            else:
                # Split the dataset into training and testing sets
                X = df.drop(target_column, axis=1)
                y = df[target_column]
                X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

                # Define the preprocessing steps for numerical and categorical features
                preprocessor = ColumnTransformer(
                    transformers=[
                        ('num', StandardScaler(), numerical_features),
                        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
                    ])
                
                # Define the pipeline with preprocessing and the classifier
                pipeline = Pipeline([
                    ('preprocessor', preprocessor),
                    ('classifier', classifier)
                ])

                if param_distributions and model_name in param_distributions:
                    param_dist = param_distributions[model_name]
                    randomized_search = RandomizedSearchCV(pipeline, param_distributions=param_dist, n_iter=n_iter,
                                                           cv=cross_validation, random_state=42, n_jobs=-1)

                    start_time = time.time()
                    randomized_search.fit(X_train, y_train)
                    time_taken = time.time() - start_time

                    best_estimator = randomized_search.best_estimator_
                    best_params = randomized_search.best_params_

                    # Make predictions
                    y_pred = best_estimator.predict(X_test)

                    # Calculate performance metrics
                    accuracy = accuracy_score(y_test, y_pred)
                    conf_matrix = confusion_matrix(y_test, y_pred)
                    class_report = classification_report(y_test, y_pred)

                    # Calculate cross-validation score
                    cv_scores = cross_val_score(best_estimator, X, y, cv=cross_validation)

                    # Store the results in the dictionary
                    model_results[model_name] = {
                        'estimator': pipeline,
                        'best_estimator': best_estimator,
                        'best_params': best_params,
                        'accuracy': accuracy,
                        'confusion_matrix': conf_matrix,
                        'classification_report': class_report,
                        'cross_val_score': cv_scores,
                        'time_taken': time_taken
                    }

                else:
                    # No randomized search, fit and evaluate with default parameters
                    start_time = time.time()
                    pipeline.fit(X_train, y_train)
                    time_taken = time.time() - start_time

                    # Make predictions
                    y_pred = pipeline.predict(X_test)

                    # Calculate performance metrics
                    accuracy = accuracy_score(y_test, y_pred)
                    conf_matrix = confusion_matrix(y_test, y_pred)
                    class_report = classification_report(y_test, y_pred)

                    # Calculate cross-validation score
                    cv_scores = cross_val_score(pipeline, X, y, cv=cross_validation)

                    # Store the results in the dictionary
                    model_results[model_name] = {
                        'estimator': pipeline,
                        'accuracy': accuracy,
                        'confusion_matrix': conf_matrix,
                        'classification_report': class_report,
                        'cross_val_score': cv_scores,
                        'time_taken': time_taken
                    }

        except Exception as e:
            print(f"Error evaluating model '{model_name}': {e}")
            model_results[model_name] = {'error': str(e)}

    return model_results
```

#### Function to Evaluate multiple machine learning models

This Function is designed to present the evaluation results of multiple machine learning models in a clear and structured format.

```python
def display_model_results(results):
    """
    Display evaluation results for multiple models.

    Parameters:
    - results (dict): Dictionary containing evaluation results for each model.

    Output:
    - Prints formatted results for each model including accuracy, cross-validation scores,
      time taken, confusion matrix, classification report, and best parameters (optional).
    """
    for model_name, result in results.items():
        if 'error' in result:
            print(f"\nError for {model_name} Model: {result['error']}")
        else:
            print(f"\nResults for {model_name} Model:")
            
            # Prepare data for the metrics table
            data = [
                ["Model", model_name],
                ["Accuracy", f"{result.get('accuracy', 'N/A') * 100:.2f} %" if 'accuracy' in result else "N/A"],
                ["Cross-Validation Scores", f"{np.mean(result.get('cross_val_score', [np.nan])):.4f}" if 'cross_val_score' in result else "N/A"],
                ["Time taken (seconds)", f"{result.get('time_taken', 'N/A'):.4f}"]
            ]
            
            # Add KMeans-specific metrics if they are available
            if model_name == 'KMeans':
                kmeans_metrics = [
                    ["Inertia", result.get('inertia', 'N/A')],
                    ["Silhouette Score", f"{result.get('silhouette_score', np.nan):.4f}" if 'silhouette_score' in result else "N/A"],
                    ["Davies-Bouldin Index", f"{result.get('davies_bouldin_index', np.nan):.4f}" if 'davies_bouldin_index' in result else "N/A"]
                ]
                data.extend(kmeans_metrics)
            
            # Print the metrics table using tabulate for a nicer display
            print(tabulate(data, headers=["Metric", "Value"], tablefmt="fancy_grid"))
            
            # Print additional details if they are available
            if 'confusion_matrix' in result and result['confusion_matrix'] is not None:
                print(f"\nConfusion Matrix:\n{result['confusion_matrix']}")
            
            if 'classification_report' in result and result['classification_report'] is not None:
                print(f"\nClassification Report:\n{result['classification_report']}")
            
            if 'best_params' in result and result['best_params'] is not None:
                print("\nBest Parameters:")
                for param, value in result['best_params'].items():
                    print(f"{param}: {value}")
            else:
                print("\nBest Parameters: N/A")
```

#### Function to visually represent the eveluation result for multiple machine learning models
This function aims to provide a visual representation of model performance across different metrics for easy comparison.

```python
def plot_model_results(results):
    """
    Plot evaluation results for multiple models based on accuracy and time taken metrics.

    Parameters:
    - results (dict): Dictionary containing evaluation results for each model.

    Output:
    - Displays bar plots for Accuracy (%) and Time Taken (seconds) metrics, highlighting the best-performing model.
    """

    # Define the metrics and their human-readable names
    metrics = ['accuracy', 'time_taken']
    metric_names = ['Accuracy (%)', 'Time Taken (seconds)']

    # Loop over each metric to create individual plots
    for metric, metric_name in zip(metrics, metric_names):
        model_names = list(results.keys())
        values = [result[metric] for result in results.values() if result.get(metric) is not None]  # Filter out None values
        
        if not values:
            print(f"No valid {metric} values found for plotting.")
            continue
        
        # Find the index of the maximum value to highlight it
        max_index = np.argmax([value if value is not None else -np.inf for value in values])

        # Create color array where the maximum value is blue, others are gray
        bar_colors = ['blue' if idx == max_index else 'gray' for idx in range(len(values))]

        # Create the plot
        plt.figure(figsize=(10, 6))
        plt.bar(model_names[:len(values)], values, color=bar_colors)  # Use model_names[:len(values)] to match lengths
        plt.xlabel('Model')
        plt.ylabel(metric_name)
        plt.title(f'{metric_name} for each model')
        plt.xticks(rotation=45)

        # Show the plot
        plt.tight_layout()
        plt.show()
```

#### Function to visually compare the model results

This function is designed to visualize and compare evaluation metrics (such as accuracy, cross-validation scores, and time taken) between models evaluated with default parameters and those tuned using RandomizedSearchCV.

```python
def plot_comparison(results_default, results_randomized):
    """
    Plot comparison of evaluation metrics between models with default parameters
    and models tuned using RandomizedSearchCV.

    Parameters:
    - results_default (dict): Dictionary containing evaluation results for models with default parameters.
    - results_randomized (dict): Dictionary containing evaluation results for models tuned with RandomizedSearchCV.

    Output:
    - Displays bar plots comparing metrics like accuracy, cross-validation scores, and time taken.
    """
    metrics_to_compare = ['accuracy', 'cross_val_score', 'time_taken']

    for metric in metrics_to_compare:
        # Extract metric scores for default and randomized search results
        default_scores = [results_default[model][metric] for model in results_default]
        randomized_scores = [results_randomized[model][metric] for model in results_randomized]

        # Plotting setup
        plt.figure(figsize=(10, 6))
        plt.bar(np.arange(len(default_scores)), default_scores, width=0.4, label='Default')
        plt.bar(np.arange(len(randomized_scores)) + 0.4, randomized_scores, width=0.4, label='Randomized')

        # Adding labels and title
        plt.xlabel('Models')
        plt.ylabel(metric.capitalize())
        plt.title(f'{metric.capitalize()} Comparison between Default and Randomized Search')

        # Setting x-axis ticks and labels
        plt.xticks(np.arange(len(default_scores)) + 0.2, results_default.keys(), rotation=45)

        # Adding legend and adjusting layout
        plt.legend()
        plt.tight_layout()

        # Display the plot
        plt.show()
```

#### Class for custom implementation of a Deep Neural Network (DNN) using Keras
This class is to provide a custom implementation of a Deep Neural Network (DNN) using Keras. It's designed for binary classification tasks with a sigmoid activation function in the output layer and binary cross-entropy loss.

```python
class KerasDNN:
    def __init__(self, input_dim=None, epochs=50, batch_size=32, learning_rate=0.001):
        self.input_dim = input_dim
        self.epochs = epochs
        self.batch_size = batch_size
        self.learning_rate = learning_rate
        self.model = self._build_model()
        
    def _build_model(self):
        inputs = Input(shape=(self.input_dim,))
        x = Dense(128, activation='relu')(inputs)
        x = Dropout(0.5)(x)
        x = Dense(64, activation='relu')(x)
        x = Dropout(0.5)(x)
        outputs = Dense(1, activation='sigmoid')(x)
        
        model = Model(inputs=inputs, outputs=outputs)
        optimizer = Adam(learning_rate=self.learning_rate)
        model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
        
        return model
    
    def fit(self, X, y):
        self.history = self.model.fit(X, y, epochs=self.epochs, batch_size=self.batch_size, verbose=0)
    
    def predict(self, X):
        proba = self.model.predict(X)
        return (proba > 0.5).astype("int32").flatten()
    
    def predict_proba(self, X):
        proba = self.model.predict(X)
        return np.hstack((1 - proba, proba))  # Return probabilities for both classes (0 and 1)
```

```python
outlier_columns = ['age', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']

# Dictionary to store outliers
outliers = {}

# Detect outliers using IQR method
for col in outlier_columns:  
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    iqr_lower_threshold = Q1 - 1.5 * IQR
    iqr_upper_threshold = Q3 + 1.5 * IQR
    iqr_outliers = df[(df[col] < iqr_lower_threshold) | (df[col] > iqr_upper_threshold)]
    outliers[f"{col}"] = iqr_outliers

# Count number of outliers
outlier_counts = {key: len(value) for key, value in outliers.items()}

# Total number of records
total_records = len(df)

# Plotting the number of outliers for each column
plt.figure(figsize=(12, 6))
bars = plt.bar(outlier_counts.keys(), outlier_counts.values(), color='skyblue')

# Annotate each bar with the percentage of outliers
for bar in bars:
    yval = bar.get_height()
    percentage = (yval / total_records) * 100
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10, f"{percentage:.2f}%", va='bottom', ha='center', fontsize=8)

plt.axhline(y=total_records, color='r', linestyle='--', label=f'Total Records: {total_records}')
plt.xlabel('Columns and Methods')
plt.ylabel('Number of Outliers')
plt.title('Number of Outliers Detected')
plt.xticks(rotation=45, ha='right')

# Place legend on top
plt.legend(loc='center right', bbox_to_anchor=(1.15, 0.5), shadow=True, ncol=1)

plt.tight_layout()
plt.show()


## Print sorted unique outliers for each column
for key, value in outliers.items():
    column_name = key.split('_')[0]
    unique_outliers = value[column_name].unique()
    sorted_unique_outliers = np.sort(unique_outliers)
    print(f"Outliers in '{key}':")
    print(sorted_unique_outliers)
    print(f"{'#' * 50}")
```

### Model Training (Hyper Parameter Tuning)

Hyperparameter tuning is a critical step in the machine learning pipeline aimed at optimizing the performance of a model by finding the best set of hyperparameters. Hyperparameters are settings that govern the learning process and the structure of a machine learning algorithm, distinct from parameters which are learned from the data during training.

#### RandomizedSearchCV
`RandomizedSearchCV` is widely recommended for hyperparameter tuning due to its efficiency, effectiveness in exploring high-dimensional spaces, and ability to balance computational resources with performance gains.
