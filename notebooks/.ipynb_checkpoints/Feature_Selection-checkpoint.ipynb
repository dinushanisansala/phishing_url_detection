## Univariate feature selection
#Univariate feature selection works by selecting the best features based on univariate statistical tests. It can be seen as a preprocessing step to an estimator

### Chi-Square Test

```python
# Split the data into features and target
X, y = split_input_data(df)

# Remove zero variance features
variance_threshold = VarianceThreshold(threshold=0)
X_nonconstant = variance_threshold.fit_transform(X)

# Get the indices of non-constant features
nonconstant_feature_indices = variance_threshold.get_support(indices=True)
nonconstant_feature_names = X.columns[nonconstant_feature_indices]

# Define the number of top features to select
k = 30  # Adjust the value of k as needed

# Check if k is less than or equal to the number of non-constant features
if k <= X_nonconstant.shape[1]:
    # Apply SelectKBest with chi2 to extract top k best features
    selector = SelectKBest(score_func=chi2, k=k)
    selector.fit(X_nonconstant, y)

    # Get the selected features
    X_select = selector.transform(X_nonconstant)

    # Get the boolean mask indicating which features are selected
    selected_features_chi2 = selector.get_support(indices=True)
    selected_feature_names_chi2 = nonconstant_feature_names[selected_features_chi2]

    print(f"Selected features: chi2 k={k}")
    for feature in selected_feature_names_chi2.tolist():
        print(feature)
else:
    print(f"Error: The number of selected features (k={k}) exceeds the number of non-constant features in X_nonconstant.")
```

###  ANOVA F-Test

```python
X, y = split_input_data(df) 

# Remove constant features to avoid warnings
constant_filter = VarianceThreshold(threshold=0)
X_nonconstant = constant_filter.fit_transform(X)

# Get the indices of non-constant features
nonconstant_feature_indices = constant_filter.get_support(indices=True)
nonconstant_feature_names = X.columns[nonconstant_feature_indices]

# Check if k is less than the number of non-constant features
k = 25  # Number of top features to select
if k <= X_nonconstant.shape[1]:
    # Apply SelectKBest class to extract top k best features
    selector = SelectKBest(score_func=f_classif, k=k)
    selector.fit(X_nonconstant, y)

    # Get the selected features
    X_select = selector.transform(X_nonconstant)

    # Get the boolean mask indicating which features are selected
    selected_features_anovaftest = selector.get_support(indices=True)
    selected_feature_names_anovaftest = nonconstant_feature_names[selected_features_anovaftest]

    print(f"Selected features: anovaftest k={k}")
    for feature in selected_feature_names_anovaftest.tolist():
        print(feature)
else:
    print(f"Error: The number of selected features (k={k}) exceeds the number of non-constant features in X_nonconstant.")
```

###  Mutual Information

```python
# Remove constant features to avoid warnings
constant_filter = VarianceThreshold(threshold=0)
X_nonconstant = constant_filter.fit_transform(X)

# Get the indices of non-constant features
nonconstant_feature_indices = constant_filter.get_support(indices=True)
nonconstant_feature_names = X.columns[nonconstant_feature_indices]

# Check if k is less than the number of non-constant features
k = 50  # Number of top features to select
if k <= X_nonconstant.shape[1]:
    # Apply SelectKBest with mutual_info_classif to extract top k best features
    selector = SelectKBest(score_func=mutual_info_classif, k=k)
    selector.fit(X_nonconstant, y)

    # Get the selected features
    X_select = selector.transform(X_nonconstant)

    # Get the boolean mask indicating which features are selected
    selected_features_mutual_info = selector.get_support(indices=True)
    selected_feature_names_mutual_info = nonconstant_feature_names[selected_features_mutual_info]

    print(f"Selected features: mutual_info_classif k={k}")
    for feature in selected_feature_names_mutual_info.tolist():
        print(feature)
else:
    print(f"Error: The number of selected features (k={k}) exceeds the number of non-constant features in X_nonconstant.")
```
