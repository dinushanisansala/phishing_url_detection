```python
def model_evaluation_with_GridSearchCV(grid_search_pipeline, X_train, X_test, y_train, y_test, feature_names, selector=True):
    """
    Evaluate SVM model using GridSearchCV and return results including best estimator, metrics, and selected features.

    Parameters:
    - grid_search_pipeline (GridSearchCV): GridSearchCV object configured with SVM pipeline and parameter grid.
    - X_train (pd.DataFrame or np.ndarray): Training features.
    - X_test (pd.DataFrame or np.ndarray): Test features.
    - y_train (pd.Series or np.ndarray): Training labels.
    - y_test (pd.Series or np.ndarray): Test labels.
    - feature_names (list): List of feature names.
    - selector (bool): Flag indicating whether to perform feature selection.

    Returns:
    - results (dict): Dictionary containing evaluation results, best estimator, parameters, and selected features.
    """

    results = {}
    start_time_pipeline = time.time()
    try:
        grid_search_pipeline.fit(X_train, y_train)
        time_taken_pipeline = time.time() - start_time_pipeline

        best_params_pipeline = grid_search_pipeline.best_params_
        best_estimator_pipeline = grid_search_pipeline.best_estimator_

        if selector:
            best_k = best_params_pipeline['selector__k']
            selector = SelectKBest(score_func=best_estimator_pipeline.named_steps['selector'].score_func, k=best_k)
            X_train_selected = selector.fit_transform(X_train, y_train)
            X_test_selected = selector.transform(X_test)
        else:
            X_train_selected = X_train
            X_test_selected = X_test

        svc = svm.SVC(C=best_params_pipeline['svc__C'], gamma=best_params_pipeline['svc__gamma'],
                      kernel=best_params_pipeline['svc__kernel'], random_state=42)
        
        start_time = time.time()
        svc.fit(X_train_selected, y_train)
        time_taken = time.time() - start_time

        y_pred = svc.predict(X_test_selected)
        accuracy = accuracy_score(y_test, y_pred)
        conf_matrix = confusion_matrix(y_test, y_pred)
        class_report = classification_report(y_test, y_pred)

        if selector:
            selected_features_indices = selector.get_support(indices=True)
            selected_feature_names = feature_names[selected_features_indices].tolist()
        else:
            selected_feature_names = feature_names.tolist()

        results = {
            'best_estimator': best_estimator_pipeline,
            'best_parameters': best_params_pipeline,
            'accuracy': accuracy,
            'confusion_matrix': conf_matrix,
            'classification_report': class_report,
            'time_taken_pipeline': time_taken_pipeline,
            'time_taken': time_taken,
            'selected_features_count': X_train_selected.shape[1] if selector else X_train.shape[1],
            'selected_features': selected_feature_names
            }
    
    except Exception as e:
        print(f"Error during model evaluation: {str(e)}")
        results = None
    
    return results
```

```python
def print_evaluation_results(results):
    if results is None:
        print("Evaluation results are not available.")
    return
    
    best_estimator = results.get('best_estimator')
    best_params = results.get('best_parameters')
    accuracy = results.get('accuracy')
    conf_matrix = results.get('confusion_matrix')
    class_report = results.get('classification_report')
    time_taken_pipeline = results.get('time_taken_pipeline')
    time_taken = results.get('time_taken')
    selected_features = results.get('selected_features')
    selected_feature_count = results.get('selected_features_count')
                    
    if best_estimator:
        print("=" * 50)
        print("Best Estimator:")
        for key, value in best_estimator.get_params().items():
            print(f"{key}: {value}")
    
    if best_params:
        print("-" * 50)
        print("Best Parameters:")
        for key, value in best_params.items():
            print(f"{key}: {value}")
    
    if accuracy is not None:
        print("-" * 50)
        print(f"Accuracy: {accuracy * 100:.2f} %")
    
    if conf_matrix is not None:
        print("-" * 50)
        print("Confusion Matrix:\n", conf_matrix)
    
    if class_report is not None:
        print("-" * 50)
        print("Classification Report:\n", class_report)
    
    if time_taken is not None:
        print("-" * 50)
        print(f"Time Taken: {time_taken} seconds")
        
    if time_taken_pipeline is not None:
        print("-" * 50)
        print(f"Time Taken Pipeline: {time_taken_pipeline} seconds")   
        
    print("=" * 50)
    if selected_feature_count is not None:
        print(f"Selected Features Count: {selected_feature_count}")
    
    if selected_features:
        print(f"Selected Features: {selected_features}")
    
    print("=" * 50)
```

```python
def print_rf_feature_importances(results, X):
    # Access the best estimator from the results
    best_estimator = results['best_estimator']
    
    # Access the RandomForestClassifier from the pipeline
    rf_classifier = best_estimator.named_steps.get('rf_classifier', None)
    
    if rf_classifier and hasattr(rf_classifier, 'feature_importances_'):
        feature_importances = rf_classifier.feature_importances_
        
        # Print feature importances
        print("Feature Importances:")
        for feature, importance in zip(X.columns, feature_importances):
            print(f"{feature}: {importance:.4f}")
    else:
        print("No feature importances available.")
```

```python
def export_model(svm_result, selector_name):
    # Get the current working directory
    current_directory = os.getcwd()

    # Define the parent directory path
    parent_directory = os.path.abspath(os.path.join(current_directory, os.pardir))

    # Create directory for saving models if it doesn't exist under the parent directory
    models_dir = os.path.join(parent_directory, "models")
    os.makedirs(models_dir, exist_ok=True)

    print(f"Directory for saving models created or already exists: {models_dir}")

    if svm_result is None or not isinstance(svm_result, dict):
        print(f"Error: svm_results is None or not a valid dictionary for selector '{selector_name}'.")
        return
    try:
        best_estimator = svm_result['best_estimator']
        best_parameters = svm_result['best_parameters']
    except KeyError as e:
        print(f"Error: Key {e} not found in svm_results for selector '{selector_name}'.")
        return
    
    model_filename = f"best_svm_model_{selector_name}.joblib"
    joblib.dump(best_estimator, os.path.join(models_dir, model_filename))

    params_filename = f"best_svm_params_{selector_name}.joblib"
    joblib.dump(best_parameters, os.path.join(models_dir, params_filename))

    if 'selected_features' in svm_result:
        selected_features_filename = f"selected_features_{selector_name}.joblib"
        joblib.dump(svm_results['selected_features'], os.path.join(models_dir, selected_features_filename))
        print(f"Selected features saved as {selected_features_filename}")

    print(f"Model saved as {model_filename}")
    print(f"Best parameters saved as {params_filename}")
```

```python
def plot_evaluation_results(svm_results):
    # Extract information from svm_results
    selector_names = list(svm_results.keys())
    accuracies = [svm_results[selector]['accuracy'] for selector in selector_names]
    time_taken = [svm_results[selector]['time_taken'] for selector in selector_names]
    num_best_features = [svm_results[selector]['selected_features_count'] for selector in selector_names]
    num_selected_features = [len(svm_results[selector]['selected_features']) for selector in selector_names]

    # Create subplots with only the necessary ones populated
    fig, axs = plt.subplots(1, 2, figsize=(14, 5))

    # Plot Accuracy and Time Taken on the first subplot (top-left)
    ax1 = axs[0]
    color_acc = 'tab:blue'
    color_time = 'tab:red'
    ax1.set_xlabel('Selector')
    ax1.set_ylabel('Accuracy', color=color_acc)
    ax1.plot(selector_names, accuracies, color=color_acc, marker='', label='Accuracy')
    ax1.tick_params(axis='y', labelcolor=color_acc)

    ax1twin = ax1.twinx()
    ax1twin.set_ylabel('Time Taken (seconds)', color=color_time)
    ax1twin.plot(selector_names, time_taken, color=color_time, marker='', label='Time Taken')
    ax1twin.tick_params(axis='y', labelcolor=color_time)

    ax1.legend(loc='upper left')
    ax1twin.legend(loc='upper right')
    ax1.set_title('Selector vs Accuracy and Time Taken')

    # Set x-axis tick positions and labels
    ax1.set_xticks(range(len(selector_names)))
    ax1.set_xticklabels(selector_names, rotation=90)

    # Plot Number of Features on the second subplot (top-right)
    ax2 = axs[1]
    ax2.set_xlabel('Selector')
    ax2.set_ylabel('Number of Features')
    ax2.plot(selector_names, num_best_features, color='tab:orange', marker='', label='Num Best Features')
    ax2.plot(selector_names, num_selected_features, color='tab:green', marker='', label='Num Selected Features')
    ax2.legend()
    ax2.set_title('Selector vs Number of Features')

    # Set x-axis tick positions and labels
    ax2.set_xticks(range(len(selector_names)))
    ax2.set_xticklabels(selector_names, rotation=90)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()

# Example usage (assuming svm_results is correctly populated)
plot_evaluation_results(svm_results)
```
