```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical

class CNNModel:
    def __init__(self, input_shape, num_classes):
        self.model = Sequential([
            Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=input_shape),
            MaxPooling2D(pool_size=(2, 2)),
            Conv2D(64, kernel_size=(3, 3), activation='relu'),
            MaxPooling2D(pool_size=(2, 2)),
            Flatten(),
            Dense(128, activation='relu'),
            Dense(num_classes, activation='softmax')
        ])
        self.model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    
    def fit(self, X_train, y_train, epochs=10, batch_size=32, validation_split=0.2):
        self.model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=validation_split, verbose=0)
    
    def predict(self, X_test):
        return np.argmax(self.model.predict(X_test), axis=1)
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.utils import to_categorical

class LSTMModel:
    def __init__(self, input_shape, num_classes):
        self.model = Sequential([
            LSTM(50, activation='relu', input_shape=input_shape),
            Dense(num_classes, activation='softmax')
        ])
        self.model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    
    def fit(self, X_train, y_train, epochs=10, batch_size=32, validation_split=0.2):
        self.model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=validation_split, verbose=0)
    
    def predict(self, X_test):
        return np.argmax(self.model.predict(X_test), axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from tensorflow.keras.utils import to_categorical
import time

def evaluate_nn_models(model_name, model, df, target_column, test_size=0.2, epochs=10, batch_size=32):
    """
    Evaluate neural network models (CNN or LSTM).

    Parameters:
    - model_name (str): The name of the machine learning model (e.g., 'CNN' or 'LSTM').
    - model: Neural network model instance.
    - df (pd.DataFrame): Input dataframe containing features and target.
    - target_column (str): Name of the target column in the dataframe.
    - test_size (float): Proportion of the dataset to include in the test split (default=0.2).
    - epochs (int): Number of epochs to train the model (default=10).
    - batch_size (int): Batch size for training (default=32).

    Returns:
    - model_result (dict): Dictionary containing evaluation results for the model.
    """
    model_result = {}  # Dictionary to store evaluation results

    # Drop non-numeric columns and handle categorical features
    df = df.dropna()
    X = df.drop(target_column, axis=1).values
    y = df[target_column].values
    
    # Encode target labels as one-hot vectors
    num_classes = len(np.unique(y))
    y = to_categorical(y, num_classes=num_classes)

    # Reshape X for CNN or LSTM
    if model_name == 'CNN':
        X = X.reshape(-1, 28, 28, 1)  # Example reshape for image data, adjust as needed
    elif model_name == 'LSTM':
        X = X.reshape(-1, 10, 10)  # Example reshape for sequential data, adjust as needed

    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    try:
        # Train the model
        start_time = time.time()
        model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.2)
        time_taken = time.time() - start_time
        
        # Make predictions
        y_pred = model.predict(X_test)
        y_pred_classes = np.argmax(y_pred, axis=1)
        y_test_classes = np.argmax(y_test, axis=1)
        
        # Calculate accuracy and confusion matrix
        accuracy = accuracy_score(y_test_classes, y_pred_classes)
        conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)

        # Store the results in the dictionary
        model_result = {
            'estimator': model,
            'accuracy': accuracy,
            'confusion_matrix': conf_matrix.tolist(),  # Convert to list for easier handling
            'time_taken': time_taken
        }

    except Exception as e:
        print(f"Error evaluating model '{model_name}': {e}")
        model_result = {'error': str(e)}

    return model_result
```
