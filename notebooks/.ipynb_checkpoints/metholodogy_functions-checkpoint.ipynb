```python
def get_stats_as_markdown_tables(df):
    """
    Calculate the descriptive statistics (min, max, mean, std) of a DataFrame grouped by the 'status' column
    and return them as separate markdown tables.

    Parameters:
    df (pd.DataFrame): The DataFrame to analyze.

    Returns:
    tuple: The descriptive statistics in markdown table format (mean, std, min, max).
    """
    # Ensure the 'status' column exists in the DataFrame
    if 'status' not in df.columns:
        raise ValueError("The DataFrame must contain a 'status' column.")

    # Group by the 'status' column and compute descriptive statistics
    grouped_stats = df.groupby('status').describe()

    # Extract the required statistics
    mean_stats = grouped_stats.loc[:, (slice(None), 'mean')]
    std_stats = grouped_stats.loc[:, (slice(None), 'std')]
    min_stats = grouped_stats.loc[:, (slice(None), 'min')]
    max_stats = grouped_stats.loc[:, (slice(None), 'max')]

    # Flatten the MultiIndex columns for better readability in the markdown table
    mean_stats.columns = ['_'.join(col).strip() for col in mean_stats.columns.values]
    std_stats.columns = ['_'.join(col).strip() for col in std_stats.columns.values]
    min_stats.columns = ['_'.join(col).strip() for col in min_stats.columns.values]
    max_stats.columns = ['_'.join(col).strip() for col in max_stats.columns.values]

    # Convert the statistics to markdown format
    mean_markdown = mean_stats.to_markdown()
    std_markdown = std_stats.to_markdown()
    min_markdown = min_stats.to_markdown()
    max_markdown = max_stats.to_markdown()

    return mean_markdown, std_markdown, min_markdown, max_markdown

```

```python
def display_stats_as_markdown_tables(mean_table, std_table, min_table, max_table):
    """
    Display the markdown tables with appropriate headers.

    Parameters:
    mean_table (str): Markdown table for mean statistics.
    std_table (str): Markdown table for standard deviation statistics.
    min_table (str): Markdown table for minimum statistics.
    max_table (str): Markdown table for maximum statistics.
    """
    display(Markdown("### Mean Table\n" + mean_table))
    display(Markdown("### Standard Deviation Table\n" + std_table))
    display(Markdown("### Minimum Table\n" + min_table))
    display(Markdown("### Maximum Table\n" + max_table))
```

```python
def plot_correlation_matrix(df, columns_to_drop=[], title=''):
    """
    Plot the correlation matrix of a DataFrame as a heatmap.

    Parameters:
    df (pd.DataFrame): The DataFrame to analyze.
    columns_to_drop (list): List of columns to drop before computing the correlation matrix.
    title (str): Title for the plot.

    Returns:
    None
    """
    # Drop specified columns and compute correlation matrix
    if columns_to_drop:
        df = df.drop(columns=columns_to_drop)
    
    correlation_matrix = df.corr()

    # Plot heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title(title)
    plt.show()
    
    return correlation_matrix
```

```python
def plot_feature_correlations(df, features, feature_category=None):
    """
    Plots the correlation of specified features with 'status' and saves the plot.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - features (list): List of feature names to check and include in the plot.
    - feature_category (str): Category name for the plot title and file saving.

    Returns:
    - None
    """
    # Ensure 'status' is in the DataFrame columns
    if 'status' not in df.columns:
        raise ValueError("'status' column must be present in the DataFrame")

    # Check if the specified features are valid columns in the DataFrame
    valid_features = [col for col in features if col in df.columns]

    # Create a DataFrame with 'status' and valid features
    if valid_features:
        length_related_df = df[['status'] + valid_features]

        # Calculate the correlation matrix
        correlation_matrix = length_related_df.corr()

        # Extract the correlation values with 'status'
        status_corr = correlation_matrix['status']

        # Drop NaN values from the correlation with 'status'
        status_corr = status_corr.dropna()

        # Sort the correlations by absolute value in descending order
        status_corr = status_corr.sort_values(ascending=False, key=abs)

        # Print the sorted correlation values with 'status'
        print(f"Sorted Correlation with 'status':")
        print(status_corr)

        # Ensure there is valid data to plot
        if status_corr.empty:
            print("No valid correlations to plot.")
            return

        # Plot the correlations
        plt.figure(figsize=(16, 8))
        barplot = sns.barplot(x=status_corr.index, y=status_corr.values, palette='coolwarm')
        plt.xticks(rotation=90)
        plt.title(f'{feature_category} - Correlation with Status' if feature_category else 'Correlation with Status')
        plt.ylabel('Correlation Coefficient')
        plt.xlabel('Features')

        # Annotate bars with correlation values
        for p in barplot.patches:
            height = p.get_height()
            x = p.get_x() + p.get_width() / 2
            y = height
            # Annotate only if height is finite
            if np.isfinite(height):
                barplot.annotate(f'{height:.2f}', 
                                 (x, y),
                                 ha='center', va='center', 
                                 xytext=(0, 5), 
                                 textcoords='offset points')

        plt.tight_layout()

        # Save and show the plot
        plt.savefig(get_file_path('output',f'{feature_category}_feature_distribution_plot.png'), bbox_inches='tight')
        plt.show()
    else:
        print("No valid features to plot.")
```

```python
def get_file_path(dir_name, file_name=None):
    """
    Constructs the full path to a file in the specified directory under the parent directory.

    Parameters:
    - dir_name (str): The name of the directory.
    - file_name (str): The name of the file (optional).

    Returns:
    - str: The full path to the file in the specified directory.
    """
    # Validate input
    if not isinstance(dir_name, str) or not dir_name:
        raise ValueError("The directory name must be a non-empty string.")

    current_directory = os.getcwd()
    # Define the parent directory path
    parent_directory = os.path.abspath(os.path.join(current_directory, os.pardir))
    # Define the target directory path
    target_directory = os.path.join(parent_directory, dir_name)

    try:
        # Create the directory if it doesn't exist
        os.makedirs(target_directory, exist_ok=True)
    except OSError as e:
        print(f"Error creating directory '{dir_name}': {e}")
        return None

    # Define the file path relative to the target directory
    file_path = os.path.join(target_directory, file_name) if file_name else target_directory
    return file_path
```

```python
def split_input_data(df):
    """
    Splits the input DataFrame into features (X) and target (y).

    Parameters:
    df (pd.DataFrame): The input DataFrame with features and target column.

    Returns:
    X (pd.DataFrame): DataFrame containing feature columns.
    y (pd.Series): Series containing the target column.
    """
    # Split the data into features and target
    X = df.drop('status', axis=1)
    y = df['status']
    return X, y
```

```python
def get_stats_group_by_status(df, url_length_features, status_column='status'):
    """
    Computes and prints statistics for URL length features overall and grouped by a status column.
    
    Parameters:
    - df (pd.DataFrame): The DataFrame containing the data.
    - url_length_features (list): List of column names to compute statistics for.
    - status_column (str): The column name used for grouping the data. Defaults to 'status'.
    
    Returns:
    - None: Prints the statistics tables.
    """
    
    # Check if the status column exists in the DataFrame
    if status_column not in df.columns:
        raise ValueError(f"Status column '{status_column}' not found in DataFrame")
    
    # Check if the URL length features exist in the DataFrame
    missing_features = [feature for feature in url_length_features if feature not in df.columns]
    if missing_features:
        raise ValueError(f"Features not found in DataFrame: {', '.join(missing_features)}")
    
    # Compute overall statistics
    overall_stats = df[url_length_features].describe().T
    
    # Reset index to use feature names as columns
    overall_stats.reset_index(inplace=True)

    print("Overall Statistics for URL Length Features:")
    display(overall_stats)
    print("\n")
    
    # Compute statistics grouped by the status column
    grouped_stats = df.groupby(status_column)[url_length_features].describe().stack(level=0).unstack(level=0)
    
    # Reshape the grouped statistics
    # Flatten the multi-index columns for readability
    grouped_stats.columns = [f'{feature}_{stat}' for feature, stat in grouped_stats.columns]
    
    # Transpose grouped statistics
    grouped_stats = grouped_stats.T
    
    # Print grouped statistics
    print(f"Statistics for URL Length Features grouped by '{status_column}':")
    display(grouped_stats)
    print("\n")
```

```python
# Function to create a DataFrame for feature importances
def create_feature_importance_df(feature_importances, feature_names):
    sorted_idx = np.argsort(feature_importances)[::-1]  # Descending order
    sorted_feature_names = np.array(feature_names)[sorted_idx]
    sorted_importances = feature_importances[sorted_idx]
    df = pd.DataFrame({
        'Feature': sorted_feature_names,
        'Importance': sorted_importances
    })
    
    return df
```

```python
def select_best_k(results_df):
    # Ensure the input is a DataFrame
    if not isinstance(results_df, pd.DataFrame):
        raise TypeError("Input must be a pandas DataFrame")

    # Ensure that necessary columns are present in the DataFrame
    required_columns = ['model', 'feature_selection_method', 'number_of_features', 'accuracy', 'time_taken']
    results_df = results_df[required_columns]
    for column in required_columns:
        if column not in results_df.columns:
            raise ValueError(f"Missing required column: {column}")

    # Sort the DataFrame by model, feature_selection_method, and accuracy in descending order
    results_sorted = results_df.sort_values(by=['model', 'feature_selection_method', 'accuracy'], ascending=[True, True, False])
    
    # Select the top 3 rows for each combination of model and feature_selection_method
    best_k_per_model_fs = results_sorted.groupby(['model', 'feature_selection_method']).head(3).reset_index(drop=True)
    
    return best_k_per_model_fs
```

```python
def plot_accuracy(df):
    """Plot Accuracy for each model."""
    if 'accuracy' in df.columns:
        plt.figure(figsize=(10, 6))
        sns.barplot(x='model_name', y='accuracy', data=df)
        plt.title('Model Accuracy')
        plt.ylabel('Accuracy')
        plt.show()
    else:
        print("No accuracy data available to plot.")

def plot_precision_recall_f1(df):
    """Plot Precision, Recall, and F1-Score for each model."""
    if 'classification_report' in df.columns:
        precision = []
        recall = []
        f1_score = []
        model_names = []
        for index, row in df.iterrows():
            if row['classification_report']:
                precision.append(row['classification_report'].get('0', {}).get('precision', 0))
                recall.append(row['classification_report'].get('0', {}).get('recall', 0))
                f1_score.append(row['classification_report'].get('0', {}).get('f1-score', 0))
                model_names.append(row['model_name'])
        
        metrics_df = pd.DataFrame({
            'model_name': model_names,
            'precision': precision,
            'recall': recall,
            'f1_score': f1_score
        })
        
        plt.figure(figsize=(10, 6))
        metrics_df.set_index('model_name').plot(kind='bar', figsize=(10, 6))
        plt.title('Precision, Recall, F1-Score by Model')
        plt.ylabel('Score')
        plt.show()
    else:
        print("No classification report data available to plot.")

def plot_time_taken(df):
    """Plot Time Taken for Training and Prediction for each model."""
    if 'time_taken_best_model_training' in df.columns and 'time_taken_best_prediction' in df.columns:
        plt.figure(figsize=(10, 6))
        df_melted = df.melt(id_vars=['model_name'], value_vars=['time_taken_best_model_training', 'time_taken_best_prediction'],
                            var_name='time_type', value_name='time')
        sns.lineplot(x='model_name', y='time', hue='time_type', data=df_melted, marker='o')
        plt.title('Time Taken for Training and Prediction')
        plt.ylabel('Time (seconds)')
        plt.show()
    else:
        print("No time data available to plot.")

def plot_number_of_features(df):
    """Plot Number of Features Selected for each model."""
    if 'number_of_features' in df.columns:
        plt.figure(figsize=(10, 6))
        sns.barplot(x='model_name', y='number_of_features', data=df)
        plt.title('Number of Features Selected')
        plt.ylabel('Number of Features')
        plt.show()
    else:
        print("No feature number data available to plot.")

def plot_confusion_matrices(df):
    """Plot Confusion Matrix for each model."""
    if 'confusion_matrix' in df.columns:
        for index, row in df.iterrows():
            conf_matrix = row['confusion_matrix']
            if conf_matrix:
                plt.figure(figsize=(6, 4))
                sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
                plt.title(f'Confusion Matrix for {row["model_name"]}')
                plt.xlabel('Predicted')
                plt.ylabel('Actual')
                plt.show()
            else:
                print(f"No confusion matrix data available for {row['model_name']}.")
    else:
        print("No confusion matrix data available to plot.")

def plot_cv_scores(df):
    """Plot Cross-Validation Scores for each model."""
    if 'cv_scores' in df.columns:
        plt.figure(figsize=(10, 6))
        for index, row in df.iterrows():
            cv_scores = row['cv_scores']
            if cv_scores:
                plt.plot(cv_scores, marker='o', label=row['model_name'])
        plt.title('Cross-Validation Scores')
        plt.xlabel('CV Iteration')
        plt.ylabel('Score')
        plt.legend()
        plt.show()
    else:
        print("No CV scores data available to plot.")

def plot_cv_scores_vs_features(df):
    """Plot Average CV Score vs Number of Selected Features."""
    if 'cv_scores' in df.columns and 'number_of_features' in df.columns:
        plt.figure(figsize=(10, 6))
        avg_scores = []
        num_features = []
        model_names = []
        for index, row in df.iterrows():
            cv_scores = row['cv_scores']
            if cv_scores:
                avg_cv_score = sum(cv_scores) / len(cv_scores)
                avg_scores.append(avg_cv_score)
                num_features.append(row['number_of_features'])
                model_names.append(row['model_name'])
        sns.scatterplot(x=num_features, y=avg_scores, hue=model_names)
        plt.title('Average CV Score vs Number of Selected Features')
        plt.xlabel('Number of Selected Features')
        plt.ylabel('Average CV Score')
        plt.legend(title='Model Name')
        plt.show()
    else:
        print("No data available to plot CV scores vs number of selected features.")

```

```python
def plot_roc_curves(df):
    """Plot ROC Curve and AUC for each model."""
    if 'roc_curve' in df.columns:
        plt.figure(figsize=(10, 6))
        labels = []
        
        for index, row in df.iterrows():
            roc_curve_data = row['roc_curve']
            
            # Debugging: Print the ROC curve data
            #print(f"Model: {row['model_name']}")
            #print(f"ROC Curve Data: {roc_curve_data}")
            
            if isinstance(roc_curve_data, tuple) and len(roc_curve_data) == 2:
                fpr, tpr = roc_curve_data
                if len(fpr) == len(tpr):
                    roc_auc = auc(fpr, tpr)
                    label = f'{row["model_name"]} (AUC = {roc_auc:.2f})'
                    plt.plot(fpr, tpr, label=label)
                    labels.append(label)
                else:
                    print(f"Length mismatch in ROC curve data for model {row['model_name']}: FPR and TPR lengths differ.")
            else:
                print(f"Incorrect ROC curve data format for model {row['model_name']}: {roc_curve_data}")
        
        if labels:
            plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
            plt.xlim([0.0, 1.0])
            plt.ylim([0.0, 1.05])
            plt.xlabel('False Positive Rate')
            plt.ylabel('True Positive Rate')
            plt.title('ROC Curve')
            plt.legend(loc='lower right')
            plt.show()
        else:
            print("No valid ROC curves to plot.")
    else:
        print("No ROC curve data available to plot.")
```

```python
def plot_cv_scores(df):
    """ Plot Cross-Validation Scores for all models. """
    fig, ax = plt.subplots(figsize=(10, 6))

    for index, row in df.iterrows():
        cv_scores = row['cv_scores']
        if cv_scores:
            ax.plot(cv_scores, marker='o', label=row['model_name'])
    
    ax.set_xlabel('CV Iteration')
    ax.set_ylabel('Score')
    ax.set_title('Cross-Validation Scores')
    ax.legend()
    plt.tight_layout()
    plt.show()
```

```python
def convert_evaluation_results_to_df(evaluation_results_final):
    """ Convert evaluation results dictionary to DataFrame. """
    models_data = []
    for model_name, model_info in evaluation_results_final.items():
        # Add model_name to the model_info
        model_info['model_name'] = model_name
        models_data.append(model_info)
    
    # Convert list of dictionaries to DataFrame
    df = pd.DataFrame(models_data)
    return df
```

```python
def plot_confusion_matrices(df):
    """ Plot Confusion Matrices for all models. """
    models_count = len(df)
    cols = 2  # Number of columns in the subplot grid
    rows = (models_count + 1) // cols  # Calculate number of rows needed
    
    fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
    axes = axes.flatten()  # Flatten the axes array for easy iteration

    for index, (model_name, row) in enumerate(df.iterrows()):
        conf_matrix = row.get('confusion_matrix', None)
        if conf_matrix is not None:
            sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', ax=axes[index])
            axes[index].set_title(f'Confusion Matrix for {row["model_name"]}')
            axes[index].set_xlabel('Predicted')
            axes[index].set_ylabel('Actual')
        else:
            axes[index].text(0.5, 0.5, f"No confusion matrix data available for {row['model_name']}.", ha='center', va='center')

    # Remove any unused subplots
    for j in range(index + 1, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()
```

```python
def visualize_model_performance(evaluation_results):
    # Convert data to a DataFrame for easier manipulation
    models_data = []
    for model_name, model_info in evaluation_results.items():
        model_info['model_name'] = model_name
        models_data.append(model_info)
    
    df = pd.DataFrame(models_data)
    
    # Create subplots with 2 columns
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 18))
    
    # Plot Accuracy
    axes[0, 0].set_title('Model Accuracy')
    if 'accuracy' in df.columns:
        sns.barplot(x='model_name', y='accuracy', data=df, ax=axes[0, 0])
        axes[0, 0].set_ylabel('Accuracy')
    else:
        axes[0, 0].text(0.5, 0.5, "No accuracy data available", ha='center', va='center')

    # Plot Precision, Recall, F1-Score
    axes[0, 1].set_title('Precision, Recall, F1-Score')
    if 'classification_report' in df.columns:
        precision = []
        recall = []
        f1_score = []
        model_names = []
        for index, row in df.iterrows():
            if row['classification_report']:
                precision.append(row['classification_report'].get('0', {}).get('precision', 0))
                recall.append(row['classification_report'].get('0', {}).get('recall', 0))
                f1_score.append(row['classification_report'].get('0', {}).get('f1-score', 0))
                model_names.append(row['model_name'])
        
        metrics_df = pd.DataFrame({
            'model_name': model_names,
            'precision': precision,
            'recall': recall,
            'f1_score': f1_score
        })
        
        metrics_df.set_index('model_name').plot(kind='bar', ax=axes[0, 1])
        axes[0, 1].set_ylabel('Score')
    else:
        axes[0, 1].text(0.5, 0.5, "No classification report data available", ha='center', va='center')

    # Plot Time Taken
    axes[1, 0].set_title('Time Taken for Search and Best Model')
    if 'time_taken_best_model_training' in df.columns and 'time_taken_best_prediction' in df.columns:
        df_melted = df.melt(id_vars=['model_name'], value_vars=['time_taken_best_model_training', 'time_taken_best_prediction'],
                        var_name='time_type', value_name='time')
        sns.barplot(x='model_name', y='time', hue='time_type', data=df_melted, ax=axes[1, 0])
        axes[1, 0].set_ylabel('Time (seconds)')
    
        # Add time values on top of the bars
        for container in axes[1, 0].containers:
            axes[1, 0].bar_label(container, fmt='%.2f', label_type='edge', rotation=90)
    else:
        axes[1, 0].text(0.5, 0.5, "No time data available", ha='center', va='center')

    # Plot Number of Features
    axes[1, 1].set_title('Number of Features Selected')
    if 'number_of_features' in df.columns:
        sns.barplot(x='model_name', y='number_of_features', data=df, ax=axes[1, 1])
        axes[1, 1].set_ylabel('Number of Features')
    else:
        axes[1, 1].text(0.5, 0.5, "No feature number data available", ha='center', va='center')
    
    plt.savefig(get_file_path('output',f'{feature_category}_feature_distribution_plot.png'), bbox_inches='tight')
    # Plot Confusion Matrices
    plot_confusion_matrices(df)

    # Plot CV Scores
    plot_cv_scores(df)
    
    # Plot Average CV Score vs Number of Selected Features
    plot_cv_scores_vs_features(df)

    # Plot ROC Curve and AUC
    plot_roc_curves(df)
```

```python
def summarize_final_results(evaluation_results_final):
    """
    Prints a summary of the final results for each model.
    
    Parameters:
    evaluation_results_final (dict): A dictionary containing the final evaluation results for each model.
    """
    print("\nSummary of Final Results:")
    
    for model_name, results in evaluation_results_final.items():
        print(f"\nModel: {model_name}")
        print(f"Model Configuration: {results['model_config']}")
        print(f"Accuracy: {results['accuracy']:.4f}")
        print(f"Time Taken for Best Model Training: {results['time_taken_best_model']:.2f} seconds")
        #print(f"Time Taken for Best Model Training: {results['time_taken_best_model_training']:.2f} seconds")
        print(f"Time Taken for Search: {results['time_taken_search']:.2f} seconds")
        print(f"Number of Features Selected: {results['number_of_features']}")
        #print(f"Features Selected: {results['selected_features']}")
        #print(f"Classifier Parameters: {results['classifier_params']}")
        print(f"Feature Selection Method: {results['feature_selection_method']}")
        print(f"Confusion Matrix:\n{np.array(results['confusion_matrix'])}")
        print(f"Classification Report:\n{results['classification_report']}")
        print(f"Cross-Validation Scores: {results['cv_scores']}")
        # Check if roc_auc is defined and has a valid value
        if 'roc_auc' in results and results['roc_auc']:
            print(f"ROC AUC: {results['roc_auc']:.4f}")
        else:
            print("ROC AUC is not available.")


```
