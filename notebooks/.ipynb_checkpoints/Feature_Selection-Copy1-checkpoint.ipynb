```python
def print_df_size(df):
    rows, columns = df.shape
    print(f"Number of rows: {rows}")
    print(f"Number of columns: {columns}")
    

def split_input_data(df):
    """
    Splits the input DataFrame into features (X) and target (y).

    Parameters:
    df (pd.DataFrame): The input DataFrame with features and target column.

    Returns:
    X (pd.DataFrame): DataFrame containing feature columns.
    y (pd.Series): Series containing the target column.
    """
    # Split the data into features and target
    X = df.drop('status', axis=1)
    y = df['status']
    return X, y
```

## Removing features with low variance

#VarianceThreshold is a simple baseline approach to feature selection. It removes all features whose variance doesn’t meet some threshold. By default, it removes all zero-variance features, i.e. features that have the same value in all samples.

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
X, y = split_input_data(df)         # Split input data into features and target

selector = VarianceThreshold()

# Fit the selector to X
X_selected = selector.fit_transform(X)

# Get the columns of X_selected that were selected
selected_columns = X.columns[selector.get_support(indices=True)].tolist()
#print("Selected Columns:", selected_columns)

# Variable DataFrame with selected columns
X = pd.DataFrame(X, columns=selected_columns)
print("Features after Removing features with low variance (zero-variance features)")
print_df_size(X)

# Function to vary v and plot number of selected columns
def plot_selected_columns_vs_variance(df):
    variance = np.arange(0, 1.05, 0.05)  # Range of variance values from 0 to 1 with step 0.05
    num_selected_columns = []  # To store number of selected columns for each v
    
    for v in variance:
        sel = VarianceThreshold(threshold=(v * (1 - v)))
        X_selected = sel.fit_transform(X)
        selected_columns = X.columns[sel.get_support(indices=True)].tolist()
        num_selected_columns.append(len(selected_columns))
    
    # Plotting number of selected columns vs v
    plt.figure(figsize=(10, 6))
    plt.plot(variance, num_selected_columns, marker='o', linestyle='-', color='b')
    plt.title('Number of Selected Columns vs Variance')
    plt.xlabel('variance')
    plt.ylabel('Number of Selected Columns')
    plt.xticks(variance)
    plt.grid(True)
    plt.show()

# Call the function to plot
plot_selected_columns_vs_variance(X)

```

```python
#Extract features for Optimal variance value
p = 0.95  

selector = VarianceThreshold(threshold=(p * (1 - p)))
X_selected = selector.fit_transform(X)

# Get the columns of X_selected that were selected
selected_feature_names_without_low_variance = X.columns[selector.get_support(indices=True)]

print("Selected features:without low variance (zero-variance features)")
for feature in selected_feature_names_without_low_variance.tolist():
    print(feature)
```

### Model Training

#### Function

```python
# Function to perform model training and evaluation with GridSearchCV
def model_evaluation_with_GridSearchCV(grid_search_pipeline, X_train, X_test, y_train, y_test):
    results = {}
    feature_importances = None
    
    # Start timing
    start_time_pipeline = time.time()

    # Fit GridSearchCV on training data
    grid_search_pipeline.fit(X_train, y_train)

    # Calculate elapsed time
    elapsed_time_pipeline = time.time() - start_time_pipeline

    # Get the best parameters and best estimator
    best_params_pipeline = grid_search_pipeline.best_params_
    best_estimator_pipeline = grid_search_pipeline.best_estimator_

    # Make predictions with the best estimator on test data
    y_pred = best_estimator_pipeline.predict(X_test)

    # Evaluate the model
    accuracy_pipeline = accuracy_score(y_test, y_pred)
    conf_matrix_pipeline = confusion_matrix(y_test, y_pred)
    class_report_pipeline = classification_report(y_test, y_pred)

    # Save the results in the dictionary
    results = {
        'best_estimator': best_estimator_pipeline,
        'best_parameters': best_params_pipeline,
        'accuracy': accuracy_pipeline,
        'confusion_matrix': conf_matrix_pipeline,
        'classification_report': class_report_pipeline,
        'time_taken': elapsed_time_pipeline
    }
    
    return results

def print_evaluation_results(results):
    best_params = results['best_parameters']
    accuracy = results['accuracy']
    conf_matrix = results['confusion_matrix']
    class_report = results['classification_report']
    time_taken = results['time_taken']
    
    print("=" * 50)
    print("Best Parameters:")
    for key, value in best_params.items():
        print(f"{key}: {value}")
    
    print("-" * 50)
    print(f"Accuracy: {accuracy * 100:.2f} %")
    print("-" * 50)
    print("Confusion Matrix:\n", conf_matrix)
    print("-" * 50)
    print("Classification Report:\n", class_report)
    print("-" * 50)
    print(f"Time Taken: {time_taken:.2f} seconds")
    print("=" * 50)

def print_rf_feature_importances(results, X):
    # Access the best estimator from the results
    best_estimator = results['best_estimator']
    
    # Access the RandomForestClassifier from the pipeline
    rf_classifier = best_estimator.named_steps.get('rf_classifier', None)
    
    if rf_classifier and hasattr(rf_classifier, 'feature_importances_'):
        feature_importances = rf_classifier.feature_importances_
        
        # Print feature importances
        print("Feature Importances:")
        for feature, importance in zip(X.columns, feature_importances):
            print(f"{feature}: {importance:.4f}")
    else:
        print("No feature importances available.")

```

### Neural Networks - Model Training

```python
# Feature scaling (recommended for neural networks)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convert labels to one-hot encoding (required for categorical_crossentropy loss)
y_train_cat = tf.keras.utils.to_categorical(y_train, num_classes=3)
y_test_cat = tf.keras.utils.to_categorical(y_test, num_classes=3)

# Define the model architecture using Input layer explicitly
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(X_train_scaled.shape[1],)),  # Input layer with input shape
    tf.keras.layers.Dense(64, activation='relu'),  # Hidden layer with 64 neurons
    tf.keras.layers.Dropout(0.2),  # Dropout layer with dropout rate 0.2
    tf.keras.layers.Dense(32, activation='relu'),  # Hidden layer with 32 neurons
    tf.keras.layers.Dense(3, activation='softmax')  # Output layer with 3 units for 3 classes in Iris dataset
])

# Compile the model
model.compile(optimizer='adam',  # Optimizer
              loss='categorical_crossentropy',  # Loss function for categorical classification
              metrics=['accuracy'])  # Evaluation metric

# Train the model
history = model.fit(X_train_scaled, y_train_cat, epochs=50, batch_size=32, validation_split=0.2, verbose=1)

# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_test_scaled, y_test_cat, verbose=0)
print(f"Accuracy: {accuracy:.2f}")

# Predictions on the test set
y_pred_proba = model.predict(X_test_scaled)
y_pred = np.argmax(y_pred_proba, axis=-1)

# Classification report and confusion matrix
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
def create_model(hidden_units=64, dropout_rate=0.2):
    model = tf.keras.models.Sequential([
        tf.keras.layers.Input(shape=(X_train_scaled.shape[1],)),
        tf.keras.layers.Dense(hidden_units, activation='relu'),
        tf.keras.layers.Dropout(dropout_rate),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(3, activation='softmax')
    ])
    
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    return model
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

# Create a KerasClassifier based on the defined create_model function
model = KerasClassifier(build_fn=create_model, verbose=0)

# Define the parameter grid for GridSearchCV
param_grid = {
    'hidden_units': [32, 64, 128],
    'dropout_rate': [0.1, 0.2, 0.3],
    'batch_size': [16, 32],
    'epochs': [50, 100]
}

# Setup GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, verbose=1, n_jobs=-1)

# Perform Grid Search to find the best parameters
grid_result = grid.fit(X_train_scaled, y_train_cat)

# Print best parameters and best accuracy found
print(f"Best Accuracy: {grid_result.best_score_:.2f}")
print(f"Best Parameters: {grid_result.best_params_}")

# Evaluate the best model on the test set
best_model = grid_result.best_estimator_
test_loss, test_accuracy = best_model.evaluate(X_test_scaled, y_test_cat, verbose=0)
print(f"Test Accuracy: {test_accuracy:.2f}")

# Predictions on the test set
y_pred_proba = best_model.predict(X_test_scaled)
y_pred = np.argmax(y_pred_proba, axis=-1)

# Classification report and confusion matrix
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

##### ML - SVM
#https://www.csie.ntu.edu.tw/~cjlin/papers/guide/guide.pdf
Scaling before applying SVM is very important. Part 2 of Sarle’s Neural Networks
FAQ Sarle (1997) explains the importance of this and most of considerations also apply to SVM. The main advantage of scaling is to avoid attributes in greater numeric
ranges dominating those in smaller numeric ranges. Another advantage is to avoid
numerical difficulties during the calculation. Because kernel values usually depend on
the inner products of feature vectors, e.g. the linear kernel and the polynomial kernel, large attribute values might cause numerical problems. We recommend linearly
scaling each attribute to the range [−1, +1] or [0, 1].

```python


# Function to plot confusion matrix
def plot_confusion_matrix(cm, classes, normalize=False, cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        title = 'Normalized confusion matrix'
    else:
        title = 'Confusion matrix'

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()

    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in np.ndindex(cm.shape):
        plt.text(j, i, format(cm[i, j], fmt),
                 ha="center", va="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
def plot_evaluation_results(results):
    kernels = list(results.keys())  # Extract kernel names from dictionary keys

    # Plotting the results
    plt.figure(figsize=(20, 5))  # Adjust the figure size to fit all subplots

    # Plot confusion matrix for each kernel
    for idx, kernel in enumerate(kernels):
        plt.subplot(1, len(kernels), idx + 1)  # Adjust subplot based on number of kernels
        plot_confusion_matrix(results[kernel]['confusion_matrix'], classes=np.unique(y), normalize=True, cmap=plt.cm.Blues)
        plt.title(f"Confusion Matrix\nKernel: {kernel}")

    plt.tight_layout()
    plt.show()
```
