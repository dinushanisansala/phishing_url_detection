```python
import os
import joblib
import json

def get_parent_or_its_sub_directory(directory=None):
    try:
        # Get the current working directory
        current_directory = os.getcwd()
        
        # Define the parent directory path
        parent_directory = os.path.abspath(os.path.join(current_directory, os.pardir))
        
        # Create directory if it doesn't exist under the parent directory
        if directory:
            directory = os.path.join(parent_directory, directory)
            os.makedirs(directory, exist_ok=True)
        
        return directory
    
    except Exception as e:
        print(f"Error creating directory: {e}")
        return None

def get_file(file, file_path=None):
    try:
        if file_path is None:
            # If file_path is not provided, use the default model_directory
            directory = get_parent_or_its_sub_directory('models')
            if directory:
                file_path = os.path.join(directory, file)
            else:
                raise FileNotFoundError("Models directory does not exist.")
        else:
            # Otherwise, check in the provided file_path
            file_path = os.path.join(file_path, file)
        
        # Check if the file path exists
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"File '{file_path}' does not exist.")
        
        # Return the full file path
        return file_path
    
    except FileNotFoundError as fnfe:
        print(f"FileNotFoundError: {fnfe}")
        return None  # Return None on error
    except Exception as e:
        print(f"Error: {e}")
        return None  # Return None on error

def load_model(file_path):
    try:
        # Load a model (joblib file)
        model = joblib.load(file_path)
        return model
    
    except FileNotFoundError as fnfe:
        print(f"FileNotFoundError: {fnfe}")
        return None  # Return None on error
    except Exception as e:
        print(f"Error: {e}")
        return None  # Return None on error

def load_features(file_path):
    try:
        # Load features (json file)
        with open(file_path, 'r') as f:
            features = json.load(f)
        return features
    
    except FileNotFoundError as fnfe:
        print(f"FileNotFoundError: {fnfe}")
        return None  # Return None on error
    except Exception as e:
        print(f"Error: {e}")
        return None  # Return None on error

selected_model_type = "RF"
selected_selector_type = "chi2"
model_file = f"{selected_model_type}_{selected_selector_type}.joblib"
model_file_path = ''  # Replace with actual model file path if needed
feature_file = f"{selected_model_type}_{selected_selector_type}_params.json"
feature_file_path = ''  # Replace with actual feature file path if needed

try:
    # Get the model file path
    model_full_file_path = get_file(model_file, model_file_path)
    feature_full_file_path = get_file(feature_file, feature_file_path)

    if model_full_file_path and feature_full_file_path:
        # Load the model
        model = load_model(model_full_file_path)
        
        # Load features
        features = load_features(feature_full_file_path)
        
        if model and features:
            print("Model and features loaded successfully.")
            # Use the loaded model and features as needed
            # Example: model.predict(features)
        else:
            print("Failed to load model or features.")
    else:
        print("Error: Failed to obtain file paths.")
        
except Exception as e:
    print(f"Error: {e}")
```
